1.	В урні знаходиться 15 білих, 5 червоних і 10 чорних куль. Випадковим чином витягується 1 куля, знайти ймовірність того, що вона буде: а) білою, б) червоною, в) чорною.

Вхідні дані:  
balls_white = 15  
balls_red = 5  
balls_black = 10  
balls_total = 15 + 10 + 5 = 30  

Розв'язок:  
a) P(balls_white) = balls_white / balls_total = 15 / 30 = 1/2  
б) P(balls_red) = balls_red / balls_total = 5 / 30 = 1/6  
в) P(balls_black) = balls_black / balls_total = 10 / 30 = 1/3  

Відповіді:  
а) 1/2  
б) 1/6  
в) 1/3

2.	В магазин надійшло 30 холодильників, п'ять з яких мають заводський дефект. Випадковим чином обирають один холодильник. Яка ймовірність того, що він буде без дефекту?

Вхідні дані:  
coolers_total = 30  
coolers_damaged = 5  
coolers_good = coolers_total - coolers_damaged = 30 - 5 = 25  

Розв'язок:  
P(coolers_good) = coolers_good / coolers_total = 25 / 30 = 5/6

Відповідь:  
5/6

3.	Набираючи номер телефону, абонент забув дві останні цифри, але пам'ятає, що одна з них — нуль, а інша — непарна. Знайти ймовірність того, що він набере правильний номер.

Вхідні дані:  
number_1 = 0  
number_2 = x, при цьому number_2 - це одне число з множини {1, 3, 5, 7, 9}

Розв'язок:  
Ймовірність того, що 0 буде вказаний на вірній позиції тобто першій або другій з двох останніх цифр становить: P(number_1) = 1/2  
Ймовірність того, що інша непарна цифра буде обрана вірно з множини варіантів становить: P(number_2) = 1/5  
Оскільки ці події мають відбутися одночасно тобто має виконатися подія 1 AND подія 2, то результуюча ймовірність того, що абонент набере правильний номер становить: P = P(number_1) * P(number_2) = 1/2 * 1/5 = 1/10

Відповідь:  
1/10

4.	Знайти ймовірність того, що при кидку двох гральних кубиків добуток чисел на них:
а) дорівнюватиме семи; б) буде не меншим за 20; в) буде парним.

Вхідні дані:  
Маємо кидок 2-х гральних кубиків, тобто загальна кількість варіантів подій: 6 * 6 = 36

Розв'язок:  
а) Оскільки 2 гральні кубики мають варіацію чисел від 1 до 6, а добуток двох чисел може дорівнювати семи лише при комбінації чисел 1 * 7 чи 7 * 1, то ймовірність P по замовчуванню буде рівна 0. Цифри 7 немає на гральних кубиках.
б) Кількість варіантів подій, коли добуток 2-х чисел від 1 до 6 буде 20 та більше становить 8.  
Перерахую дані комібнації: 4 * 5, 5 * 4, 4 * 6, 6 * 4, 5 * 6, 6 * 5, 5 * 5, 6 * 6.  
Відповідно тоді P = 8/36 = 2/9   
в) Щоб добуток двох чисел був парним достатньо, щоб одне з чисел було парним. Тому:  
- якщо на першому кубику випаде парне число з множини {2, 4, 6}, то на другому кубику може випасти будь-яке число з множини {1, 2, 3, 4, 5, 6}, відповідно це 3 * 6 = 18 варіантів;
- якщо на першому кубику випаде непарне число з множини {1, 3, 5}, то на другому кубику для виконання умови мають випасти виключно парні числа тобто число з множини {2, 4, 6}, тобто це 3 * 3 = 9 варіантів.  

Звідси загальна кількість варіантів для виконання умови становить 18 + 9 = 27 варіантів.  
Отже, шукана ймовірність складає P = 27 / 36 = 3/4

Відповіді:  
а) 0  
б) 2/9  
в) 3/4  

5.	Ймовірність того, що на тесті з біології учень О. правильно розв'яже більше 11 задач, дорівнює 0,67. Ймовірність того, що О. правильно розв'яже більше 10 задач, дорівнює 0,74. Знайдіть ймовірність того, що О. правильно розв'яже рівно 11 задач.

Вхідні дані:    
P(tasks > 11) = 0.67, відповідно P(tasks <= 11) = 1 - 0.67 = 0.33  
P(tasks > 10) = 0.74  

Розв'язок:  
Рівно 11 задач підходить під умову Ймовірністі того, що О. правильно розв'яже більше 10 задач, при цьому одначно це буде доповненням до Ймовірності того, що О. правильно розв'яже більше 11 задач.  
Дані 2 умови мають виконатися одночасно тобто шукана P = P(tasks > 10) * P(tasks <= 11) = 0.74 * 0.33 = 0.2442

Відповідь:
0.2442
