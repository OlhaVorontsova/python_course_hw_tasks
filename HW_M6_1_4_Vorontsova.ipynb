Домашнє завдання №4

1.	З колоди у 52 карти випадково витягується одна карта. Яка ймовірність того, що це туз, якщо відомо, що ця карта — пікова?

Вхідні дані:  
Колода на 52 карти  
Колода включає 4 масті, тобто по 13 карт кожної масті  
Колода включає 4 тузи  

Розв'язок:  
Ймовірність витягнути піковий туз становить P(AB) = 1/52  
Ймовірність витягнути пікову карту становить P(B) = 13/52  
Тоді умовна ймовірність витягнути туз, якщо карта пікова становить: P(A|B) = 1/52 : 13/52 = 1/13  
Відповідь: 1/13

2.	У групі з 30 студентів є 18 хлопців і 12 дівчат. Відомо, що 7 хлопців і 5 дівчат носять окуляри. Випадково обраний студент носить окуляри. Яка ймовірність того, що цей студент — дівчина?

Вхідні дані:  
Група включає 30 студентів  
В групі хлопців 18  
В групі дівчат 12  
В групі носять окуляри 12 студентів, з них 5 дівчат та 7 хлопців

Розв'язок:  
Ймовірність вибору дівчини в окулярах P(AB) = 5/30  
Ймовірність вибору студента в окулярах P(B) = 12/30  
Тоді умовна ймовірність вибору дівчини серед студентів в окулярах P(A|B) = 5/30 : 12/30 = 5/12  
Відповідь: 5/12

3.	У конверті знаходяться 10 лотерейних білетів, серед яких 3 виграшних. З конверта послідовно витягуються білети. Знайти ймовірності того, що:
а) 2-й витягнутий білет буде виграшним, якщо 1-й був виграшним; б) 3-й буде виграшним, якщо попередні два білети були виграшними; в) 4-й буде виграшним, якщо попередні білети були виграшними.

Вхідні дані:  
Всього лотерейних білетів 10  
Виграшних білетів 3

Розв'язок:  
а) Якщо 1-й витягнутий білет був виграшним, то залишилося 2 виграшних білети з 9 доступних. Отже, P(A|B) = 2/9  
Відповідь: 2/9  
б) Якщо перші 2 білети були виграшні, то залишився 1 виграшний білет серед 8 доступних. Отже P(A|B) = 1/8  
Відповідь: 1/8  
в) Якщо перші 3 білети витягнуті виграшні, то виграшні білети закінчилися і умовна ймовірність витягнути 4-й виграшний білет рівна 0.  
Відповідь: 0
